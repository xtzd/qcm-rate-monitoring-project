```python
import pandas as pd
```

```python
data = pd.read_csv('RUN#06237.txt',engine='python',sep='\t',header=0,names=['Timestamp','Status','Rate','Thick_1','Thick_0','Frequency'])
```

```python
data.Timestamp = pd.to_datetime(data.Timestamp)
```

```python
import matplotlib.pyplot as plt
```

```python
x = (data['Timestamp']-data['Timestamp'][0]).dt.total_seconds().tolist()
```

```python
y=data.Thick_0.tolist()
```

```python
plt.rcParams['lines.markersize']=1
plt.scatter(x[6550:7000],y[6550:7000],marker='x',linewidths=0)
plt.ylabel('Rate (A/s)')
plt.xlabel('Time (s)')
plt.show()
```

```python
plt.axis([0, 1000, 0, 1])
plt.ion()

start = 100
while True:
    xx = x[start:start+500]
    yy = y[start:start+500]
    plt.cla()
    plt.scatter(xx, yy)
    plt.pause(0.033)
    start +=10
```
